## v. Analisis dan Perhitungan

### Tren Penjualan Tiap Bulan

```python
# Mengubah tipe data `Date` dari string menjadi datetime
df['Date'] = pd.to_datetime(df['Date'])

# Mengubah data `Date` menjadi index
df_index = df.set_index('Date')

# Gabungan total penjumlahan penjualan tiap bulan
tren_penjualan = df_index['Amount ($)'].resample('ME').sum()

tren_penjualan.plot(marker='o', figsize=(10, 5), title='Tren Penjualan Tiap Bulan')
plt.ylabel('Total Amount ($)')
plt.xlabel('Month')
plt.show()
```

**Insight**

Dapat dilihat penjualan paling tinggi terjadi di bulan Juli dengan sekitar $32000, kemudian diikuti di bulan Maret dan bulan Juni. Sedangkan penjualan paling rendah terjadi di bulan April dengan sekitar kurang dari $18000

### Total Unit Produk Yang Terjual Tiap Brand

```python
unit_per_brand = df.groupby('Brand')['Quantity'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=unit_per_brand.index, y=unit_per_brand.values)
plt.title('Total Unit Produk Yang Terjual Tiap Brand')
plt.xlabel('Brand')
plt.ylabel('Total Unit')
plt.show()
```

```python
df.groupby('Brand')['Quantity'].sum().value_counts()
```

**Insight**

Dapat dilihat bahwa brand Nike menjadi brand yang unit produk-nya paling banyak terjual di angka sekitar 365 unit, diikuti brand Adidas di angka 210 unit dan brand Off-White di angka mendekati 143 unit. Sedangkan untuk brand Essentials menjadi brand dengan unit produk yang terjual paling sedikit dengan angka kurang dari 95 unit.

```python
df.groupby('Country')['Amount ($)'].sum().value_counts()
```

**Insight**

Dapat dilihat bahwa negara Jepang berada di posisi pertama dalam menghasilkan nilai penjualan, kemudian diikuti Canada di posisi kedua, lalu diikuti juga oleh Germany dan UK Country di posisi ketiga dan keempat. Sedangkan untuk negara India berada di posisi terakhir.

### Perbedaan Signifikan Harga Pada Tiap Jenis Produk

- Untuk jenis produk akan menggunakan tipe produk sneakers dan hoodie karena dua tipe produk tersebut mempunyai penjualan jumlah unit paling tinggi.

```python
## ANOVA
sneakers = df['Unit Price ($)'][df['Product Type']=='Sneakers']
hoodie = df['Unit Price ($)'][df['Product Type']=='Hoodie']
```

```python
print(f'Rata-rata harga unit Sneakers: {sneakers.mean():.2f}')
print(f'Rata-rata harga unit Hoodie: {hoodie.mean():.2f}')
```

```python
res = stats.f_oneway(sneakers, hoodie)
print('P-value:',res.pvalue)
```

**Insight**

Jika diasumsikan
H0: Tidak ada perbedaan harga pada tipe produk Sneakers dan Hoodie
H1: Terdapat perbedaan harga pada tipe produk Sneakers dan Hoodie

maka H0 ditolak karena nilai p-value kurang dari 0.05 yang artinya memang ada perbedaan harga pada produk sneakers dan hoodie. Namun, meskipun ada perbedaan harga tetap membuat dua produk tersebut menjadi yang paling banyak diminati.
